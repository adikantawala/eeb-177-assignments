```python
in_file = open("cocoli.txt")
#read in first line to skip over header
in_file.readline()
#set up empty dict

sizes_94 = {}
sizes_97= {}
sizes_98= {}

for line in in_file:
    items = line.split()
    tree_id = items[0]
    size_in_94 = float(items[4])
    size_in_97 = float(items[5])
    size_in_98 = float(items[6])
    
    sizes_94[tree_id]= size_in_94
    sizes_97[tree_id]= size_in_94
    sizes_98[tree_id]= size_in_94
```

```python
#(list(sizes_97.values()) - list(sizes_94.values())/list(sizes_94.values())
```

```python
import matplotlib.pyplot as plt

#plot some histograms 
#matplotlib works best with list

sizes_in_94 = list(sizes_94.values())
plt.hist(sizes_in_94, log =True)
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(1,2,sharex = True)
ax1.hist(list(sizes_in_94), log = False)
ax1.set_title("unlogged")
ax2.hist(sizes_in_94,log = True)
ax2.set_title("logged")
plt.show()
```
